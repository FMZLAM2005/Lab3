# Exploratory Data Analysis on Student Performance Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
print("The libraries were successfully loaded")
```

```python
df = pd.read_csv("student_performance.csv")
df.head()
```

```python
print("Shape:", df.shape)
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isna().sum())
```

The dataset contains 15 rows and 5 columns. There are no missing values. The dataset includes both numerical and categorical variables.

```python
df.describe()
```

The average final grade is around 80. Students show variation in study hours and attendance.

```python
plt.figure(figsize=(6,4))
sns.histplot(df["FinalGrade"], bins=5)
plt.title("Distribution of Final Grades")
plt.show()
```

The distribution shows that most students scored between 70 and 95.

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="StudyHours", y="FinalGrade", data=df)
plt.title("Study Hours vs Final Grade")
plt.show()
```

There is a positive relationship between study hours and final grade.

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="Attendance", y="FinalGrade", data=df)
plt.title("Attendance vs Final Grade")
plt.show()
```

Higher attendance appears to be associated with higher final grades.

```python
plt.figure(figsize=(5,4))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

Final grade has strong positive correlation with study hours, attendance, and previous grades.

Final Conclusion

This exploratory data analysis shows that study hours and attendance significantly influence student performance. Students who dedicate more time to studying and maintain higher attendance tend to achieve better final grades. The dataset does not contain missing values, making it suitable for further modeling.
